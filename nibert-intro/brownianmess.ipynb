```python
from itertools import cycle
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
dim=3
step_n=5 #number of movements we're tracking
step_set=[-1, 0, 1]
origin=np.zeros((1,dim)) #start point of particle, 1x3 array, the one means one row array w x,y,z

step_shape=(step_n,dim)
steps=np.random.choice(step_set,size=step_shape) #randomly mixing numbers (-1,0,1) from step_set into size of array and we want size of array to be step_shape
path1=np.concatenate([origin, steps]) #creates one array starting at origin and with of all the steps(-1,0,1) (the path of particle)
path=path1.cumsum(0) #final position of particle, cumulatively adds the columns of origin+steps
start=path[:1] #first element. start (which is 0) to end 1 (which is 1st index, second element) first array of path [start:end]
stop=path[-1:] #from start position but use -1 bc moves back 1 from the start which is the end array of path

print(stop)
path
```

```python
# Plot the path
fig = plt.figure(figsize=(5,5),dpi=200) #dpi is resolution (dots per inch) doesn't really matter
ax = fig.add_subplot(111, projection='3d') #the 111 is location of plot. ex 223 would mean there is 2X2 of graphs (4 graphs) and we want third of the 4 graphs
ax.grid(False)
ax.xaxis.pane.fill = ax.yaxis.pane.fill = ax.zaxis.pane.fill = False
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.scatter3D(path[:,0], path[:,1], path[:,2], c='blue', alpha=0.25,s=1) #x,y,z is 0th,1st,2nd element of cumsum path
ax.plot3D(path[:,0], path[:,1], path[:,2], c='blue', alpha=0.5, lw=0.5) #path traveled. syntax is [:,number] means all the rows in column specified (x,y,z)
ax.plot3D(start[:,0], start[:,1], start[:,2], c='red', marker='*') #start marker 
ax.plot3D(stop[:,0], stop[:,1], stop[:,2], c='black', marker='o') #end marker
plt.title('one random walk')
#plt.savefig(‘plots/random_walk_3d.png’,dpi=250);
```
