```python
import matplotlib.pyplot as plt
import matplotlib.dates as md
import locale 
```

```python
#want to list dates in chronological order
xlist=[]
for x in x_val: #x is one of the arrays of the whole entire array x_val ex: whole May 2021 array
    isnewarray=True #is it start of new array? yes, year is start of new array
    for xdata in x: #an indivual element in the single array x ex: 2021-may, 05/07...
        if isnewarray: 
            year=xdata[0:4] #only keeps first 4 digits of first element in array
        else:
            xlist.append(xdata+'/'+year) #no longer start of array so add year to each element in array
        isnewarray=False 
print(xlist)

convertx=md.num2date(md.datestr2num(xlist)) #convert from string (str2num), then convert num to date

ylist=[]
locale.setlocale(locale.LC_ALL,'en_US.UTF-8') #converting to integers in a string with commas to int
for y in y_val:
    for ydata in y:
         ylist.append(locale.atoi(ydata))
print(ylist)
```
