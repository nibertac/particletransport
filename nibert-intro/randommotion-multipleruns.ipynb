```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt

from itertools import cycle
from mpl_toolkits.mplot3d import Axes3D
colors=cycle('bgrcmykbgrcmykbgrcmykbgrcmyk') 
```

```python
dims=3
n_part=40 #number of particles
step_n=20
step_set=[-1, 0 ,1]
runs=np.arange(n_part) #make number of particles into an array
step_shape=(step_n,dims) #7x3 array
# Plot
fig=plt.figure(figsize=(10,10),dpi=250)
ax=fig.add_subplot(111, projection='3d')
ax.grid(False)
ax.xaxis.pane.fill=ax.yaxis.pane.fill=ax.zaxis.pane.fill=False
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
 
for i, col in zip(runs, colors):
    #Simulate steps in 3D
    origin=np.random.randint(low=-10,high=10,size=(1,dims)) #choose numbers lower than -10 or higher than 10
    steps=np.random.choice(a=step_set, size=step_shape)
    path=np.concatenate([origin, steps]).cumsum(0)
    start=path[:1]
    stop=path[-1:]
    #Plot the path
    ax.scatter3D(path[:,0], path[:,1], path[:,2],
                 c=col,alpha=0.15,s=.05);
    ax.plot3D(path[:,0], path[:,1], path[:,2], 
              c=col, alpha=0.5,lw=1.25, ls='--')
    ax.plot3D(start[:,0], start[:,1], start[:,2],
              c=col, marker='*')
    ax.plot3D(stop[:,0], stop[:,1], stop[:,2],
              c=col, marker='o');
plt.title('3D Random Walk - Multiple runs')
print(origin)
```
